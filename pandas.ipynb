<h1><b>Merge</b></h1>

```python
data1=pd.DataFrame({"a":[4,3,6,7,8],"b":[6,3,1,8,4]})
data2=pd.DataFrame({"a":[4,3,6,7,8],"c":[66,23,21,18,54]})
pd.merge(data1,data2,on="a")
```

```python
#in above data if i change the value of any of the one data of a then this merge fucntion will only give the merge of common data only
```

```python
data3=pd.DataFrame({"a":[4,3,6,7,8],"b":[6,3,1,8,4]})
data4=pd.DataFrame({"a":[4,3,6,7,9],"c":[66,23,21,18,54]})
pd.merge(data3,data4,how="right")#left=data3,right=data4
```

```python
pd.merge(data3,data4,how="outer",indicator=True)#indicator parameter helps us to check the merge function
```

```python
pd.merge(data3,data4)
```

```python
#if the data is not showing then write this as written below:
#pd.merge(data1,data2,left_index=True,right_index=True)
```

```python
pd.merge(data3,data4,left_index=True,right_index=True,suffixes=("Name","id"))
```

<h1><b>Concat</b></h1>

```python
series1=pd.Series([2,3,4,5])
series2=pd.Series([9,8,7,6])
pd.concat([series1,series2])
```

```python
data=pd.DataFrame({"a":[2,3,4,6],"b":[7,6,9,8]})
data2=pd.DataFrame({"a":[1,4,3,5],"b":[3,9,5,2]})
pd.concat([data,data2])
```

```python
data3=pd.DataFrame({"a":[2,3,4,6],"b":[7,6,9,8]})
data4=pd.DataFrame({"a":[1,4,3,5],"b":[3,9,5,2]})
pd.concat([data3,data4],axis=1)
```

```python
data3=pd.DataFrame({"a":[2,3,4,6],"b":[7,6,9,8]})
data4=pd.DataFrame({"a":[1,4],"b":[5,2]})
pd.concat([data3,data4],axis=1,join="inner")#inner ->intersection whereas outer->union
```

<h1><b>JOIN AND APPEND FUNCTION IN PYTHON</b></h1>

<h1><b>JOIN</b></h1>

```python
data3=pd.DataFrame({"a":[2,3,4,6],"b":[7,6,9,8]})
data4=pd.DataFrame({"c":[11,42,45],"d":[52,25,64]})
data3.join(data4)
```

```python
data4=pd.DataFrame({"a":[2,3,4,6],"b":[7,6,9,8]})
data5=pd.DataFrame({"c":[11,42,45],"d":[52,25,64]})
data4.join(data5,how="inner")
```

```python
data4=pd.DataFrame({"a":[2,3,4,6],"b":[7,6,9,8]})
data5=pd.DataFrame({"b":[11,42,45],"c":[52,25,64]})
data4.join(data5,how="inner",lsuffix="_123",rsuffix="_2")#rsuffix->right suffix wehereas lsuffix->left suffix
```

<h1><b>APPEND</b></h1>

```python
data4=pd.DataFrame({"a":[2,3,4,6],"b":[7,6,9,8]})
data5=pd.DataFrame({"b":[11,42,45],"c":[52,25,64]})
data4.append(data5)
```

<h1><b>Pivot</b></h1>
